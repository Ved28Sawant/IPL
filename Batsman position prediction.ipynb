```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Batsman standings.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
x=df.drop(['Position','Player','Balls faced','Runs','Strength','strike rate','Average'],axis=1)
y=df['Position']
```

```python
x
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.3)
from sklearn.linear_model import LinearRegression as LR
lr=LR()

model=lr.fit(x_train,y_train)
```

```python
predict=lr.predict(x_test)
print("predicted positions:",predict)
df2=pd.DataFrame(x_test,y_test)
df2
```

```python
Position=model.predict(np.array([[78,70,10,80,0,5,50,100]]))
print("Position:",Position)
```

```python
model.score(x_test,y_test)
```
