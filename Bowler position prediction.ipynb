```python
x=df.drop(['Strike rate','4 wickets in an overs','5 wickets in an over','Average','Position','Player','Runs','Economy'],axis=1)
y=df['Position']
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

model=lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
Position=lr.predict([[120,120,200,130]])
print("Predicted position in standings:",Position)
```
